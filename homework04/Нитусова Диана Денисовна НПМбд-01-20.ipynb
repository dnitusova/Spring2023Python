Аналитическое решение интеграла №2$: \frac{e - \sqrt[4]{e}}{2} \approx 0.71712820$

Аналитическое решение интеграла №9$: \frac{\pi}{2^\frac{3}{2}} \approx 1.11072073$

```python
import math
h2 = 0.00001
h9 = 0.0000001
m = int((2 - 3 / 4) / h9)
xx = np.arange(1, 2, h2)
x = np.append(xx, 2)
y = np.linspace(3 / 4, 2, m, endpoint=False)

def func2(x):
    return (math.e ** (1 / x ** 2) / x ** 3)

def func9(x):
    return 1 / np.sqrt(2 + 3 * x - 2 * x ** 2, where = x!=2)

rect2 = h2 * np.sum(func2(x))
trape2 = h2 / 2 * (func2(x[0]) + 2 * np.sum(func2(x[1:-1])) + func2(x[-1]))
simp2 = h2 / 3 * (func2(x[0]) + 2 * np.sum(func2(x[2:-1:2])) + 4 * np.sum(func2(x[1:-2:2])) + func2(x[-1]))

rect9 = h9 * np.sum(func9(y))
trape9 = h9 / 2 * (func9(y[0]) + 2 * np.sum(func9(y[1:-1])) + func9(y[-1]))
simp9 = h9 / 3 * (func9(y[0]) + 2 * np.sum(func9(y[2:-1:2])) + 4 * np.sum(func9(y[1:-2:2])) + func9(y[-1]))

print('Метод прямоугольников №2 =', rect2, '\nМетод трапеций №2 =', trape2, '\nПравило Симпсона №2 =', simp2)
print('\nМетод прямоугольников №9 =', rect9, '\nМетод трапеций №9 =', trape9, '\nПравило Симпсона №9 =', simp9)
print('\nВывод: С помощью численных методов можно посчитать интеграл с заданной точностью.')
print('       Чем меньше шаг интегрирования, тем точнее решение.')
```
