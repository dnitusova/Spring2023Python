3. В Scipy также есть много менее распространенных специальных функций. Взгляните на обширный список <a href=https://docs.scipy.org/doc/scipy/reference/special.html> здесь</a>!
```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import scipy.special as sc
```

### Пункт 1.

```python
xx = np.linspace(-1, 1, 250)
yy = xx ** 2 * np.sin(1 / xx ** 2) + xx
plt.title(u'График', family='verdana')
plt.xlabel(u'Ось x', family='verdana')
plt.ylabel(u'Ось y', family='verdana')
plt.plot(xx, yy)
```

### Пункт 2.

```python
xx = np.linspace(5, 25, 75)
aa = 1 / (1 + xx ** 2)
bb = 1 / xx ** 2
yy = abs(1 - aa / bb)
plt.title(u'График', family='verdana')
plt.xlabel(u'Ось x', family='verdana')
plt.ylabel(u'Ось y', family='verdana')
plt.semilogy(xx, yy)
```

### Пункт 3.

```python
xx = np.linspace(-1, 1, 250)
yy = sc.gammasgn(xx)
yy2 = sc.erf(xx)
plt.title(u'График', family='verdana')
plt.xlabel(u'Ось x', family='verdana')
plt.ylabel(u'Ось y', family='verdana')
plt.plot(xx, yy, label = 'Sign of the gamma function.', color = 'green')
plt.plot(xx, yy2, label = 'Returns the error function of complex argument.', color = 'red')
plt.legend()
```

## Exercise 2  
1. Построить график функции $sin(x)/x$ на отрезке $[-6\pi;6\pi]$ и добавить тики по оси $x$ с шагом $\pi/2$ от $-6\pi$ до $6\pi$, причём обозначения тиков должны содержать греческую букву $\pi$. Добавить сетку на график.

```python
xx = np.linspace(- 6 * np.pi, 6 * np.pi, 250)
yy = np.sin(xx) / xx
xt = np.arange(- 6 * np.pi, 6 * np.pi + np.pi / 2, np.pi / 2)
yt = []
for i in range(- 12, 13):
    if i % 2 == 0:
        if i == 0:
            yt.append('0')
        elif i == - 2:
            yt.append(r'$-\pi$')
        elif i == 2:
            yt.append(r'$\pi$')
        else:
            yt.append(r'$%s\pi$'%(i // 2))
    else:
        if i != 1 and i != - 1:
            yt.append(r'$\dfrac{%s\pi}{2}$'%(i))
        elif i == - 1:
            yt.append(r'$-\dfrac{\pi}{2}$')
        else:
            yt.append(r'$\dfrac{\pi}{2}$')
            
plt.figure(figsize = (14, 7))
plt.title(u'График', family='verdana')
plt.xlabel(u'Ось x', family='verdana')
plt.ylabel(u'Ось y', family='verdana')
plt.plot(xx, yy)
plt.xticks(xt, yt)
plt.grid()
```

## Exercise 3  
1. Создайте цветовой график $sin(x)sin(y)$ на интервале $[−π,π]×[−π,π]$.
2. Создайте цветовой график $e^{-x^2-y^2}$ на интервале $[−3,3]×[−3,3]$.

### Пункт 1.

```python
x = np.linspace(- np.pi, np.pi, 250)
y = np.linspace(- np.pi, np.pi, 250)
X,Y = np.meshgrid(x,y)

Z = np.sin(X) * np.sin(Y)

fig, ax = plt.subplots(figsize = (6,5))

norm = mpl.colors.Normalize(- abs(Z).max(), abs(Z).max())

p = ax.pcolor(X, Y, Z, norm = norm, cmap = mpl.cm.twilight)
ax.axis('tight')
ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(4))

cb = fig.colorbar(p, ax = ax)
cb.set_ticks([-1, -.5, 0, .5, 1])
```

### Пункт 2.

```python
x = np.linspace(- 3, 3, 250)
y = np.linspace(- 3, 3, 250)
X,Y = np.meshgrid(x,y)

Z = np.exp(- X ** 2 - Y ** 2)

fig, ax = plt.subplots(figsize = (6,5))

norm = mpl.colors.Normalize(- abs(Z).max(), abs(Z).max())

p = ax.pcolor(X, Y, Z, norm = norm, cmap = mpl.cm.Pastel1)
ax.axis('tight')
ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(4))

cb = fig.colorbar(p, ax = ax)
cb.set_ticks([-1, -.5, 0, .5, 1])
```
