```python
import numpy as np
import numpy.linalg
import numpy.random
import matplotlib.pyplot as plt
```

### Полиномиальная регрессия.

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(- 1 / 2 * x)

def approximation(n):
    a = np.linspace(1, 15, n)
    A = np.array([[x ** i for i in range(n)] for x in a])
    B = f(a)
    W = np.linalg.solve(A, B)
    answer = np.array([sum([W[i] * x ** i for i in range(n)]) for x in a])
    return answer

xx = np.linspace(1, 15, 250)
n = 5
xx2 = np.linspace(1, 15, n)
plt.grid()
plt.title(u'Исходная функция', family='verdana')
plt.xlabel(u'Ось x', family='verdana')
plt.ylabel(u'Ось y', family='verdana')
plt.plot(xx, f(xx))
plt.plot(xx2, f(xx2), marker = '.', linestyle = '', label = f'Точки для аппроксимации при n = {n}')
plt.legend()
plt.show()

plt.grid()
plt.title(u'Аппроксимация', family='verdana')
plt.xlabel(u'Ось x', family='verdana')
plt.ylabel(u'Ось y', family='verdana')
plt.plot(xx2, approximation(n))
plt.plot(xx2, f(xx2), marker = '.', linestyle = '', label = f'Точки для аппроксимации при n = {n}')
plt.legend()
plt.show()
```

### Полиномиальная регрессия с зашумлённым данными.

```python
def noize_approximation(n):
    a = np.linspace(1, 15, n)
    A = np.array([[x ** i for i in range(n)] for x in a])
    B = f(a)
    for i in range(len(B)):
        B[i] += np.random.normal(0, 0.01)
    W = np.linalg.solve(A, B)
    answer = np.array([sum([W[i] * x ** i for i in range(n)]) for x in a])
    return answer, B


xx = np.linspace(1, 15, 250)
n = 5
xx2 = np.linspace(1, 15, n)
yy = noize_approximation(n)
plt.grid()
plt.title(u'Исходная функция', family='verdana')
plt.xlabel(u'Ось x', family='verdana')
plt.ylabel(u'Ось y', family='verdana')
plt.plot(xx, f(xx))
plt.plot(xx2, yy[1], marker = '.', linestyle = '', label = f'Точки для аппроксимации при n = {n}')
plt.legend()
plt.show()

plt.grid()
plt.title(u'Аппроксимация', family='verdana')
plt.xlabel(u'Ось x', family='verdana')
plt.ylabel(u'Ось y', family='verdana')
plt.plot(xx2, yy[0])
plt.plot(xx2, yy[1], marker = '.', linestyle = '', label = f'Точки для аппроксимации при n = {n}')
plt.legend()
plt.show()
```

### Метод наименьших квадратов .

```python
def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-1/2 * x)

def least_squares(n):
    N = 100
    a = np.linspace(1, 15, N)
    A = np.array([[x ** i for i in range(n)] for x in a])
    B = f(a)
    for i in range(len(B)):
        B[i] += np.random.normal(0, 0.1)
    W, unnes1, unnes2, unnes3 = np.linalg.lstsq(A, B, rcond=None)
    answer = np.array([sum([W[i] * x**i for i in range(n)]) for x in a])
    return answer, B

xx = np.linspace(1, 15, 250)
N = 100
xx2 = np.linspace(1, 15, N)
n = 4
yy = least_squares(n)

plt.grid()
plt.title(u'Исходная функция', family='verdana')
plt.xlabel(u'Ось x', family='verdana')
plt.ylabel(u'Ось y', family='verdana')
plt.plot(xx, f(xx))
plt.plot(xx2, yy[1], marker = '.', linestyle = '', label = f'Точки для аппроксимации при n = {n}')
plt.legend()
plt.show()

plt.grid()
plt.title(u'Аппроксимация', family='verdana')
plt.xlabel(u'Ось x', family='verdana')
plt.ylabel(u'Ось y', family='verdana')
plt.plot(xx2, yy[0], label = f'Аппроксимация при n = {n}')
plt.plot(xx2, yy[1], marker = '.', linestyle = '', label = f'Точки для аппроксимации при n = {n}')
plt.legend()
plt.show()
```
